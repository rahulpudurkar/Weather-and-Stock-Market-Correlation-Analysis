```python
# Assignment Part 2 begins here
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import plotly
plotly.offline.init_notebook_mode(connected=True)
from plotly.graph_objs import *
from plotly import tools
import plotly.graph_objects as go
import seaborn as sns
import plotly.express as px
import pandas as pd
```

```python
# QUESTION 1

# Load datasets and parse date column
df_weather = pd.read_csv("./london_weather.csv", parse_dates = ['date'])
df_stock = pd.read_csv("./Index closing price from 1994 to 2021.csv", parse_dates = ['Date'])
```

```python
df_weather.head()
```

```python
df_stock.head()
```

```python
# Used plotly express to create a hoverable line graph plotting the date vs ftse index value over time.
fig = px.line(df_stock, x='Date', y='ftse', labels={'ftse': 'FTSE Value'}, title='FTSE Index Over Time')

# Updating the properties to show hover content.
fig.update_layout(hovermode='x unified', hoverlabel=dict(bgcolor="white", font_size=12))


fig.show()
```

```python
max_value_row = df_stock.loc[df_stock['ftse'].idxmax()]

print(max_value_row)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 12), sharex=True)

indices = ['spx', 'dax', 'ftse', 'nikkei']
for i, index in enumerate(indices):
    axes[i].plot(df_stock['Date'], df_stock[index], label=index)
    axes[i].set_ylabel(f'{index} Index')
    axes[i].grid(True)
    axes[i].legend()

fig.suptitle('Stock Indices Over Time', fontsize=16)

plt.show()
```

```python
# PLOT HEATMAP OF CORRELATION BETWEEN ALL PAIRS OF indices.
correl = df_stock.corr()

trace = go.Heatmap(z=correl.values,
                  x=correl.index.values,
                  y=correl.columns.values)
data=[trace]
plotly.offline.iplot(data, filename='basic-heatmap')
```

```python
# Filter data for the years 2005 to 2010
df_filtered = df_stock[df_stock['Date'].dt.year.between(2005, 2010)]

fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 12), sharex=True)

indices = ['spx', 'dax', 'ftse', 'nikkei']
for i, index in enumerate(indices):
    axes[i].plot(df_filtered['Date'], df_filtered[index], label=index)
    axes[i].set_ylabel(f'{index} Index')
    axes[i].grid(True)
    axes[i].legend()

fig.suptitle('Stock Indices Over Time (2005-2010)', fontsize=16)

plt.show()
```

```python
stock_dataset_filtered = df_stock[df_stock['Date'].dt.year.between(2005, 2010)]

correl_filtered = stock_dataset_filtered.corr()

trace = go.Heatmap(z=correl_filtered.values,
                   x=correl_filtered.index.values,
                   y=correl_filtered.columns.values)

layout = go.Layout(title='Correlation Matrix (2005-2010)',
                   xaxis=dict(title='Stock Indices'),
                   yaxis=dict(title='Stock Indices'))

fig = go.Figure(data=[trace], layout=layout)

fig.show()
```

```python
# QUESTION 2

weather_data_filtered = df_weather[df_weather['date'].dt.year.between(2014, 2018)]
weather_data_filtered.head()
```

```python
weather_data_filtered.isna().sum()
```

```python
stock_data_filtered = df_stock[df_stock['Date'].dt.year.between(2014, 2018)]
stock_data_filtered.head()
```

```python
stock_data_filtered.isna().sum()
```

```python
#Interpolating the columns with NaN values
weather_data_filtered['global_radiation'].interpolate(inplace = True)
weather_data_filtered['snow_depth'].interpolate(inplace = True)
```

```python
weather_data_filtered.isna().sum()
```

```python
stock_data_filtered.isna().sum()
```

```python
len(weather_data_filtered)
```

```python
len(stock_data_filtered)
```

```python
stock_subset = stock_data_filtered[['Date', 'ftse']]

# Merge the columns with the London weather data using the Date field, used left_on and right_on as 
# the two date columns have different starting letters.
merged_data = pd.merge(weather_data_filtered,stock_subset, left_on='date', right_on='Date', how='left')
```

```python
merged_data.head()
```

```python
len(merged_data)
```

```python
merged_data.isna().sum()
```

```python
merged_data['ftse'].interpolate(inplace=True)
```

```python
merged_data.drop(columns=['Date'], inplace=True)
```

```python
merged_data.isna().sum()
```

```python
# filtering data where snow depth is greater than 0
subset_data = merged_data[merged_data['snow_depth'] > 0]
subset_data.head()
```

```python
correl = subset_data.corr()

trace = go.Heatmap(z=correl.values,
                  x=correl.index.values,
                  y=correl.columns.values)
data=[trace]
plotly.offline.iplot(data, filename='basic-heatmap')
```
